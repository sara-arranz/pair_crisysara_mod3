 Verifica si hay valores nulos en el conjunto de datos.

df.isna() y df.isnull() son funcionalmente equivalentes en pandas. 

Ambas funciones se utilizan para identificar los valores faltantes (NaN) en un DataFrame o una Serie.

isna() es un alias de isnull().

Ambas devuelven un DataFrame o Serie de booleanos, donde True indica que el valor es NaN (valor faltante) y False indica que el valor no es NaN.

Por lo tanto, puedes usar cualquiera de las dos según tu preferencia, ya que ambas hacen exactamente lo mismo.

```python
df_eda.isnull()
```

```python
df_eda.isna().sum()
```

Si queremos acceder al valor nulo en concreto de una posición, miramos el DataFrame generado de booleanos e iteramos
