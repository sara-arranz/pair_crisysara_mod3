### Selección de Método de Imputación: 

Discute con tu compañera cuál sería la mejor estrategia para manejar los valores nulos en cada una de las columnas identificadas en el paso anterior.

```python
columnas_categoricas = df_nulos.select_dtypes(include='O')
columnas_categoricas
```

```python
# Obtenemos la lista de columnas categóricas que tienen nulos
nulos_categoricas = df_nulos[df_nulos.columns[df_nulos.isnull().any()]].select_dtypes(include = "O").columns
print("Las columnas categóricas que tienen nulos son : \n ")
print(nulos_categoricas)
```

```python
# sacamos el 'value_counts()'  de cada una de las columnas categóricas que tienen nulos para saber como es la distribución de sus categorías
for col in nulos_categoricas:
    print(f"La distribución de las categorías para la columna {col.upper()}")
    display(df_nulos[col].value_counts() / df_nulos.shape[0])
    print("........................")
```

Tras valoración de los datos de cada columna categórica hemos decidido que tipo de método usar con los valores nulos: 

En una lista añadiremos las columnas en las que vamos a utilizar el método unk y en la otra en la que utilizaremos la moda

```python
columnas_datos_desconocidos= ['abbreviation', 'capital/major_city', 'currency-code', 'largest_city','continent']
columna_datos_moda = 'official_language'
```

```python
for col in columnas_datos_desconocidos:
    df_nulos[col] = df_nulos[col].fillna("Unknown")
```

Generamos la moda de la única columna que hemos decidido importar con la moda

```python
moda_language = df_nulos['official_language'].mode()[0]
moda_language
```

La aplicamos. No usamos bucle porque solo tenemos una columna

```python
df_nulos['official_language'] = df_nulos['official_language'].fillna(moda_language)
```

```python
df_nulos[columnas_datos_desconocidos].isnull().sum()
```

```python
df_nulos[columna_datos_moda].isnull().sum()
```

#### REVISIÓN VALORES CONTINUOS

```python
columnas_medianas = ['agricultural_land', 'birth_rate', 'cpi', 'fertility_rate', 'forested_area',
 'gross_primary_education_enrollment', 'gross_tertiary_education_enrollment',
 'life_expectancy', 'out_of_pocket_health_expenditure',
 'physicians_per_thousand', 'population_labor_force_participation',
 'tax_revenue', 'latitude', 'longitude']
```

Aplicamos la mediana en las columnas numéricas con valores que no tienen gran diferencia entre la media y la mediana.

```python
df_nulos[columnas_medianas] = df_nulos[columnas_medianas].fillna(df_nulos[columnas_medianas].median())
```

```python
df_nulos[columnas_medianas].describe().T
```
