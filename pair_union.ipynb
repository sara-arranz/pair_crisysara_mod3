Las columnas unemployment_rate, total_tax_rate, tax_revenue, population_labor_force_participation, out_of_pocket_health_expenditure, gross_tertiary_education_enrollment, gross_primary_education_enrollment, forested_area, cpi_change, agricultural_land tienen "%". Elimina los "%" de los valores de las columnas.

```python
df_union.columns
```

```python
# Quitar los guiones bajos al final de los nombres de las columnas
df_union.columns = df_union.columns.str.rstrip('_')
```

```python
columnas_modificar = ['unemployment_rate', 'total_tax_rate', 'tax_revenue', 'population:_labor_force_participation', 'out_of_pocket_health_expenditure', 'gross_tertiary_education_enrollment','gross_primary_education_enrollment', 'forested_area', 'cpi_change', 'agricultural_land']
```

```python
for column in columnas_modificar:
    df_union[column] = df_union[column].str.replace('%', '', regex=False)  # Eliminar '%'
```

```python
df_union[columnas_modificar].head(3)
```

Haz lo mismo para las columnas de gasoline_price, gdp, minimum_wage, pero eliminando "$".

```python
columnas_modificar2 = ['gasoline_price', 'gdp', 'minimum_wage']
```

```python
for column in columnas_modificar2:
    df_union[column] = df_union[column].str.replace('$', '', regex=False)  # Eliminar '$'
```

### Ejercicios de Filtrado

```python
df_union.columns
```

Encuentra todos los países cuya mortalidad infantil esté entre 40 y 50 personas por kilómetro cuadrado.

```python
mortalidad_infantil = df_union[(df_union['infant_mortality'] >= 40) & (df_union['infant_mortality'] <= 50)]
resultado = mortalidad_infantil[['country','infant_mortality']]
display(resultado)
```

Encuentra los países cuyas tasas de natalidad son mayores o iguales a 20 y su esperanza de vida es mayor de 75 años.

```python
natalidad = df_union[(df_union['birth_rate'] >= 20) & (df_union['life_expectancy'] > 75)]
resultado2= natalidad[['country','birth_rate','life_expectancy']]
display(resultado2)
```

Encuentra las ciudades cuyos paises contienen la palabra "la" en su nombre.

```python
paises_filtrado = df_union[(df_union['country'].str.contains('la'))]
resultado3 = paises_filtrado[['country', 'largest_city']]
display(resultado3)
```

Encuentra los países cuyos medicos por cada 1000 habitantes (physicians_per_thousand) sea mayores de 5.

```python
medicos_filtrados = df_union[df_union['physicians_per_thousand'] > 5]
resultado4 = medicos_filtrados[['country','physicians_per_thousand']]
display(resultado4)
```
